```python
train_datagen = ImageDataGenerator(
    # rescale=1./255, 
                                   rotation_range=40, 
                                   width_shift_range=0.2, height_shift_range=0.2, 
                                   horizontal_flip=True, fill_mode='nearest', zoom_range= 0.2, 
                                   preprocessing_function=preprocess_input)
test_datagen = ImageDataGenerator(
    # rescale=1./255,
                                  preprocessing_function=preprocess_input)
val_datagen = ImageDataGenerator(
    # rescale=1./255,
                                 preprocessing_function=preprocess_input)
```

```python
train_data = train_datagen.flow_from_dataframe(train, x_col='data', y_col='labels', color_mode='rgb', class_mode='categorical', target_size=(120, 120), shuffle=False)
test_data = test_datagen.flow_from_dataframe(test, x_col='data', y_col='labels', color_mode='rgb', class_mode='categorical', target_size=(120, 120),  shuffle=False)
val_data = val_datagen.flow_from_dataframe(val, x_col='data', y_col='labels', color_mode='rgb', class_mode='categorical', target_size=(120, 120), shuffle=False)
```

```python
fig, axes = plt.subplots(4, 8, figsize = (20, 10))
for i, ax in enumerate(axes.flatten()):
    ax.set_title(np.argmax(train_data[0][1][i]))
    ax.imshow(train_data[0][0][i])
    ax.axis('off')


plt.show()
```

```python
def build_model(train_data, val_data):
    Base = ResNet50(include_top=False, input_shape=(120, 120, 3))
    for layer in Base.layers[:-10]:
        layer.trainable = False
        
        
    model = Sequential()
    model.add(Base)
    model.add(GlobalAveragePooling2D())
    model.add(Flatten())
    model.add(BatchNormalization())
    model.add(Dropout(.3)) 
    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(.3))
    model.add(Dense(102, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    histpry = model.fit(
        train_data,
        steps_per_epoch=len(train_data),
        epochs=115, validation_data=(val_data))
    
    
    # conv_base.trainable = True
    
#     model = Sequential()
    
#     model.add(Base)
#     model.add(GlobalAveragePooling2D())
#     model.add(Flatten())
    
#     model.add(BatchNormalization())

    
#     model.add(Dense(1024, activation='relu',
#                     # kernel_regularizer=regularizers.l1_l2(l1=0.01, l2=0.01)
#              ))
#     model.add(Dropout(0.2))
    
    
    
#     model.add(Dense(256, activation='relu', 
#                     # kernel_regularizer=regularizers.l1_l2(l1=0.01, l2=0.01)
#              ))
#     model.add(Dropout(0.2))
    
    
#     model.add(Dense(102, activation='softmax'))
#     model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', 'Recall', 'Precision'])
    
#     history = model.fit(train_data, validation_data=val_data, batch_size=32, epochs=200,
#                        # callbacks=tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=30, restore_best_weights=True)
#                        )
#     print(history)
    return model, history, model.summary() 
```

```python
def artifacts(model, history, test_data):
    y_pred = np.argmax(model.predict(test_data), axis=1)
    y_true = test_data.classes
    
    
    conf = ConfusionMatrixDisplay(confusion_matrix(y_true, y_pred))
    conf.plot(cmap='Blues')
    plt.title('Confusion Matrix')
    plt.savefig('images/Confusion_Matrix')
    plt.show()
    plt.close()
    
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.savefig('images/Accuracy')
    plt.show()
    plt.close()
  
    # Plot the training and validation loss over time
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.savefig('images/Loss')
    plt.show()
    plt.close()
```

```python
mlflow.set_tracking_uri('http://localhost:5000')
mlflow.set_experiment(experiment_name='ResNet50')
mlflow.start_run(run_name=f"run_{next(x)}")
mlflow.tensorflow.autolog()

model, history, summary =  build_model(train_data, val_data)
artifacts(model, history, test_data)

mlflow.log_artifacts('images/')
```
